> 2. A __*state of activation*__
> 3. An __*output function*__ for each unit
> 4. A __*pattern of connectivity*__ among units

5. What were the two theoretical misunderstandings that held back the field of neural networks?

> In 1969, Marvin Minsky and Seymour Papert wrote the book "Perceptrons" which showed that a single layer of artificial neurons was unable to learn mathematical functions, like the XOR logic gate. While they subsequently demonstrated in the same book that additional layers can solve this problem, only the first insight was recognized, leading to the start of the first AI winter.
> 
> In the 1980’s, models with two layers were being explored. Theoretically, it is possible to approximate any mathematical function using two layers of artificial neurons. However, in practice, these networks were too big and too slow. While it was demonstrated that adding additional layers improved performance, this insight was not acknowledged, and the second AI winter began. In this past decade, with increased data availability, and improvements in computer hardware (both in CPU performance but more importantly in GPU performance), neural networks are finally living up to its potential.

6. What is a GPU?

> GPU stands for Graphics Processing Unit (also known as a graphics card) and it's responsible for making a lot of computations. CPUs, or central processing units, are the core units of all standard computers, and they execute the instructions that make up computer programs. GPUs, on the other hand, are specialized units meant for displaying graphics, especially the 3D graphics in modern computer games. The hardware optimizations used in GPUs allow it to handle thousands of tasks at the same time. Incidentally, these optimizations allow us to run and train neural networks hundreds of times faster than a regular CPU.

10. Why is it hard to use a traditional computer program to recognize images in a photo?

> For us humans, it is easy to identify different objects in photos, or to identify the main label of a photo, such as identifying cats vs dogs in a photo. This is because, subconciously, our brains have learned the underlying patterns and features that define a cat or a dog for example. However, it's hard to imagine how we could define such rules for a computer program to recognize a cat or a dog. How would you program that? Can you think of a universal rule to determine if a photo contains a cat or dog? This is very difficult because cats, dogs, and other objects, have a wide variety of shapes, textures, colors, and other features. It's close to impossible to manually program this in a way that a computer would be able to recognize those patterns.

11. What did Samuel mean by "weight assignment"?

> In 1949, Arthur Samuel, an IBM researcher started working on a different way to get computers to complete tasks, which he called "machine learning". In 1962, he published the classic essay "Artificial Intelligence: A Frontier of Automation" where the basic idea was instead of telling the computer the exact steps required to solve a problem, show it examples of problems to solve, and let it figure it out how to solve it itself. This turned out to be very effective, and in 1961, his checkers-playing program beat the Connecticut state champion, the first time a computer beat a human-being in a game of intelligence.
>
> His idea to build intelligent computers involves __*weight assignments*__. Weights are just variables, and a weight assignment is choosing values for those variables. Those weights assignments (values) are combined with inputs, which are other set of values (e.g., image pixels), and together they define how the program will operate. 
>
>Arthur Samuel further mentions an __*"automatic means of testing the effectiveness of any current weight assignment"*__ and a __*"mechanism for altering the weight assignment so as to maximize the performance"*__. This refers to the evaluation and training of the model in order to obtain a set of parameter values that maximizes model performance.

14. Why is it hard to understand why a deep learning model makes a particular prediction?

> A neural network is a particular kind of deep learning model, which is a subset of machine learning. The adjective "deep" from deep learning refers to many layers in the network. Those networks can have hundreds of layers, where each neuron in one layer is affecting the connected neurons in subsequent layers. Altogether, due to the complex nature of deep learning models, it is very difficult to understand why a neural network makes a given prediction.
>
> However, recent research has made it easier to better understand neural network’s predictions. For example, we can determine what kind of features activate neurons in the network by analyzing the sets of weights.  Also, when applying CNNs to images, it's possible to see which parts of the images are highly activated by the model.

15. What is the name of the theorem that shows that a neural network can solve any mathematical problem to any level of accuracy?

> __*Universal approximation theorem*__ is a mathematical proof that shows a neural network can solve any problem to any level of accuracy, in theory.

16. What do you need in order to train a model?

> You need data related to the task you want to solve, and an architecture chosen for that given task. For most use-cases of deep learning, you will need labels for your data to compare your model predictions. You will need a loss function that will quantitatively measure the performance of your model. And you need a way to update the parameters of the model in order to improve its performance (this is known as an optimizer). 

17. How could a feedback loop impact the rollout of a predictive policing model?

> A model is as good as the data used to train it. If we use biased data to train the model, the model will produce biased predictions. In a feedback loop, the more the model is used, the more biased the data becomes, making the model even more biased and so forth.
>
> For example, we may want a model that would predict crimes, but we use information on arrests as a proxy . However, this data itself is slightly biased due to the biases in existing policing processes. Training with this data leads to a biased model. Law enforcement might use the model to determine where to focus police activity, increasing arrests in those areas. These additional arrests would be used in training future iterations of models, leading to an even more biased model. This cycle continues as a positive feedback loop.

18. Do we always have to use 224×224-pixel images with the cat recognition model?

> It's a standard size used by convention for historical reasons. For example, old pretrained models required this size exactly. If you increase the size you will often get better results, since the model will be able to focus on more details, but it's burdensome for speed and memory consumption. The opposite is true if size is decreased.

19. What is the difference between classification and regression?

> Classification focuses on making predictions about a class, or category. It's predicting from a number of discrete possibilities, such as "dog" or "cat". Regression focuses on making predictions about numerical quantities, such as temperature or house prices.

20. What is a validation set? What is a test set? Why do we need them?

> In machine learning, we use a portion of the data as the __*training set*__ and another portion as the __*validation set*__. The training set is used to __*train*__ the model, and the validation set is used for __*evaluating*__ the performance of the model during training, in order to prevent overfitting. We don't want our model to get good results by "cheating" or memorizing the patterns in the training set, but rather because it learns the appropriate features to use for prediction on new unseen data.
>
> Moreover, another portion of the model is used as the __*test set*__, which measures the __*accuracy*__ of the model. The accuracy of the model comes only from the test set, and never the validation set. Splitting the data is necessary because we care about how well our model __*generalizes*__ on __*unseen data*__.

21. What will fastai do if you don't provide a validation set?

> fastai defaults __*valid_pct*__ to $0.2$, so even if you forget, fastai will create a validation set for you.

22. Can we always use a random sample for a validation set? Why or why not?

> A good validation or test set should be representative of new data you will see in the future. Sometimes this isn’t true if a random sample is used. For example, for time series data, selecting sets randomly does not make sense. Instead, defining different time periods for the train, validation, and test set is a better approach.

23. What is overfitting? Provide an example.

> __*Overfitting*__ is when a model trains for too long, or there's not enough data, the model learns too well the patterns in the training set rather than generalizing the underlying patterns in the data.

24. What is a metric? How does it differ from "loss"?

> A __*metric*__ is a function used to measure the quality of the model's predictions using the validation set. In this case, we're using __*error_rate*__ as the metric. In a classification model, the error_rate tells what percentage of images in the validation set are being classified incorrectly.
>
> A __*loss*__, on the other hand, defines a "measure of performance", that is used to update the weights automatically during the training of the model. A good choice for loss is a choice that is easy for stochastic gradient descent to use. But a metric is defined for human consumption, so a good metrci is one that is easy for you to understand.

25. How can pretrained models help?

> A model that already has weights already trained on another dataset is called a __*pretrained*__ model. Because the model has been already trained on a dataset, it already learned many of the underlying patterns in the data, thus, it's already very capable of making predictions. Using pretrained models is the __*most*__ important method we have that allow us to train more accurate models, more quickly, with less data, and less time and money.
>
> __*Transfer learning*__ is the method for using pretrained models in a different task from what they were originally trained for.

26. What is the "head" of a model?

> A pretrained model will remove the last layer of the network and replace it with one or more new layers with randomized weights of an appropriate size for the dataset you are working with. This last part of the model is known as the __*head*__. The head of a model is the part that is newly added to be specific to the new dataset.

27. What kinds of features do the early layers of a CNN find? How about the later layers?

> The first layers of a CNN model recognize edges,  lines, circles, colors, and other simple patterns of an image.

28. Are image models only useful for photos?

> A CNN is the model typically used for image recognition tasks. However, a CNN is not only useful for recognizing images. It can be used for a variety of tasks, such as, sound classification, time series analysis, malware classification, etc.

29. What is an "architecture"?

> The __*template*__ of the model that we're trying to fit, i.e., the actual mathematical function that we're passing the input data and parameters to.

30. What is segmentation?

> A model that can recognize the content of every individual pixel in an image is called __*segmentation*__.

31. What is `y_range` used for? When do we need it?

> __*y_range*__ is used to tell the range of the target the model is trying to predict. It is used when predicting continuous numbers (e.g., predicting movie ratings, range of 0.5-5).

32. What are "hyperparameters"?

> __*Hyperparameters*__ are parameters that define __*how*__ the model is trained. Those modeling choices are the network architecture, learning rates, data augmentation strategies, and other factors. 

33. What's the best way to avoid failures when using AI in an organization?

> Key things to consider when using AI in an organization:
> 1. Make sure a training, validation, and testing set is defined properly in order to evaluate the model in an appropriate manner.
> 2. If you're considering bringing in an external vendor, make sure that you hold out some test data that the vendor _never gets to see_.
> 3. Try out a simple baseline, which future models should hopefully beat. Or even this simple baseline may be enough in some cases.
